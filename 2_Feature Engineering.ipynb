# 2 - Feature Engineering

In this module, we will try to create new Features which will be helpful to cluster/group the customers more efficiently.

```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns 
import os
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
os.chdir(r'D:\OneDrive - Northeastern University\Jupyter Notebook\Data Science Projects\Bank-Customer-Profiling-and-Segmentation')
df = pd.read_csv('./data/marketing_data_clean.csv')
df.head()
```

```python
df['BALANCE_USAGE_RATIO'] = df['BALANCE'] / df['CREDIT_LIMIT']
df['ONEOFF_PURCHASE_RATIO'] = df['ONEOFF_PURCHASES'] / (df['PURCHASES'] + 1e-6)
df['INSTALLMENT_PURCHASE_RATIO'] = df['INSTALLMENTS_PURCHASES'] / (df['PURCHASES'] + 1e-6)
df['TOTAL_PURCHASES'] = df['ONEOFF_PURCHASES'] + df['INSTALLMENTS_PURCHASES']
df['PURCHASES_PER_TRX'] = df['PURCHASES'] / df['PURCHASES_TRX']
df['CASH_ADVANCE_PER_TRX'] = df['CASH_ADVANCE'] / df['CASH_ADVANCE_TRX']
df['PAYMENT_RATIO'] = df['PAYMENTS'] / df['CREDIT_LIMIT']
df['MINIMUM_PAYMENT_RATIO'] = df['MINIMUM_PAYMENTS'] / df['CREDIT_LIMIT']
df['FULL_PAYMENT_FLAG'] = (df['PRC_FULL_PAYMENT'] == 1).astype(int)
df['UTILIZATION_RATIO'] = (df['PURCHASES'] + df['CASH_ADVANCE']) / df['CREDIT_LIMIT']
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Checking missing variable
df.isnull().sum().sort_values(ascending=False).head()
```

```python
df.fillna(0, inplace=True)
```

```python
df.to_csv('./data/marketing_data_clean_w_features.csv', index = False)
```
